Data Preprocessing (Cleaning & Transformation)

```python
# Check for missing values
df.isnull().sum()
```

```python
# Check for duplicated values
df.duplicated().sum()
```

```python
# Convert Year column to datetime format
df['Year'] = pd.to_datetime(df['Year'], format='%Y')
```

```python
df['Profit_Margin_%'] = (df['Net Income'] / df['Revenue']) * 100
df['Profit_Margin_%'] = df['Profit_Margin_%'].fillna(0)  # Replace NaN with 0
```

```python
df.sort_values(by=['Company ', 'Year'], inplace=True)  # Ensure correct ordering

df['Revenue_Growth_%'] = df.groupby('Company ')['Revenue'].pct_change() * 100
df['Revenue_Growth_%'] = df['Revenue_Growth_%'].fillna(0)
```
